```python
# Node structure
class Node:
    def __init__(self, x):
        self.data = x
        self.next = None


# Stack implementation using Linked List
class mystack:
    def __init__(self):
        self.top = None
        self.count = 0

    # Push operation
    def push(self, x):
        temp = Node(x)
        temp.next = self.top
        self.top = temp
        self.count += 1

    # Pop operation
    def pop(self):
        if self.top is None:
            print("Stack Underflow")
            return -1

        temp = self.top
        self.top = self.top.next
        self.count -= 1
        return temp.data

    # Peek operation
    def peek(self):
        if self.top is None:
            print("Stack is empty")
            return -1
        return self.top.data

    # Check if stack is empty
    def isEmpty(self):
        return self.top is None

    # Size of stack
    def size(self):
        return self.count


# Main program
st = mystack()

# Get number of elements
n = int(input("Enter the number of elements to push: "))

# Push elements
for i in range(n):
    x = int(input(f"Enter element {i + 1}: "))
    st.push(x)

# Pop one element
print("\nPopped Element:", st.pop())

# Peek
print("Top element:", st.peek())

# Check if stack is empty
if st.isEmpty():
    print("Stack is empty")
else:
    print("Stack is not empty")

# Size
print("Current size:", st.size())


```
